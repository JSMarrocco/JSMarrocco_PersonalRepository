```python
###############################################
# Description:
#   Intro to Deep learning #1
#   from Siraj Raval
#
# Author: Jean-Simon Marrocco
# Date: 29-11-2017
################################################

import pandas as pd                     #let us read are data set
from sklearn import linear_model        #Machine learning library for this exempel 
import matplotlib.pyplot as plt        #Visual model and data
import tensorflow as tf

#read data
dataframe = pd.read_fwf('brain_body.txt')
x_values = dataframe [['Brain']]
y_values = dataframe [['Body']]


#We gonna use Linear Regression to predict the body weight

#trains mdoel on data
body_reg = linear_model.LinearRegression()
body_reg.fit(x_values, y_values)

#visualize results
plt.scatter(x_values, y_values)
plt.plot(x_values, body_reg.predict(x_values))
plt.show()
```
